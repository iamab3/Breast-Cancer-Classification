```python
# Deep Learning model for Breast Cancer Classification with Neural Networks
# importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import sklearn.datasets
```

```python
# Loading dataset from sklearn
Breastcancer_data = sklearn.datasets.load_breast_cancer()
print(Breastcancer_data)
```

```python
# Convert the data to pandas dataframe
Cancer_df = pd.DataFrame(Breastcancer_data.data, columns=Breastcancer_data.feature_names)
Cancer_df.head()
```

```python
# Loading the target variable to the dataset
Cancer_df['Target'] = Breastcancer_data.target
Cancer_df.tail()
```

```python
# # of rows and columns
Cancer_df.shape
```

```python
# Getting information on data
Cancer_df.info()
```

```python
# Checking for null values
Cancer_df.isnull().sum()
```

```python
# Checking statistics of the data
Cancer_df.describe()
```

```python
# Checking the number of values of Target, 1 -> Benign, 0-> Malignant(this cancer spreads)
Cancer_df['Target'].value_counts()
```

```python
# Since the difference in datapoints between two types of cancer is not too much, we can get good fitting with this data.
# Checking the mean of each feature corresponding to target values
Cancer_df.groupby('Target').mean()
```

```python
# Splitting data into features and target
X = Cancer_df.drop(columns='Target', axis=1)
Y = Cancer_df['Target']
```

```python
X.head()
```

```python
print(Y)
```

```python
# Splitting into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
# Checking the split in data
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Standardise the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

```python
# This is to check whether the values are in similar ranges or not after scaling
print(X_train_std)
```

```python
# importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers on neural network
# Where 20, 10 and 2 refer to number of nodes at each step. At the last step, we have 2 nodes because the output can be either 0 or 1 so equal to number of classes
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(10, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# Compiling the neural network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Training the neural network on unstandardised data. Check the accuracy here is 85% (Changes as we refresh it more)
history1 = model.fit(X_train, Y_train, validation_split=0.1, epochs=10)
```

```python
# Training the neural network. Check the accuracy here is 96%
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

```python
# Visualizing accuracy and loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation data'], loc='lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data','validation data'], loc='upper right')
plt.show()
```

```python
# Accuracy of the model on test data
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy*100)
```

```python
print(X_test_std.shape)
print(X_test_std[0])
```

```python
# Predicting now
Y_pred = model.predict(X_test_std)
print(Y_pred.shape)
print(Y_pred[0])
```

```python
# model.predict gives the prediction probability for each class for that data point

# Converting prediction probability to class labels
Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)

# 0-> first is maxx, 1-> second is maxx
```

```python
# Building the predictive system
input_data = ()

# Change input data to numpy array
numpy_input = np.asarray(input_data)

# reshaping the numpy array as we are predicting for only one data point
reshaped_input = numpy_input.reshape(1,-1)

# Standardizing the input data
input_data_std = scaler.transform(reshaped_input)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label =[np.argmax(prediction)]
print(prediction_label)

if prediction_label[0]==1:
    print("The tumour is benign")
else:
    print("The tumour is malignant")
```
